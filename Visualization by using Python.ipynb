## 4. Boxplot

### Some Statistics Info :-

1. Measure of Central Tendency : Mean | Median

2. Measures of Variability : Range (Max minus Min) | IQR (Interquartile Range) | Std-dev | Var

```python
sns.boxplot(df_phi['S_Poro']);
```

```python
sns.boxplot(df_phi['S_Poro']);
plt.axvline(df_phi['S_Poro'].quantile(0.5), color='red', alpha=0.5)
```

```python
IQR = df_phi['S_Poro'].quantile(0.75) - df_phi['S_Poro'].quantile(0.25)
IQR
```

```python
sns.boxplot(df_phi['S_Poro']);
plt.axvline(df_phi['S_Poro'].quantile(0.5), color='red', alpha=0.5)
plt.axvline(df_phi['S_Poro'].quantile(0.25), color='red', alpha=0.5)
plt.axvline(df_phi['S_Poro'].quantile(0.75), color='red', alpha=0.5)

plt.axvline(df_phi['S_Poro'].quantile(0.50) + 1.5*IQR, color='green', alpha=0.5)
plt.axvline(df_phi['S_Poro'].quantile(0.50) - 1.5*IQR, color='green', alpha=0.5)
plt.grid()
```

## 5. Subplots - Multiple plots next to each other

```python
fig , axs = plt.subplots(nrows=2, ncols=2);
fig.set_figheight(6) ; fig.set_figwidth(10)
x_data = np.linspace(1,100,1000)

axs[0,0].plot(x_data, x_data**1, label='Power = 1')
axs[0,0].set_title('Power = 1')
axs[0,0].grid()

axs[0,1].plot(x_data, x_data**2, label='Power = 2', color='red')
axs[0,1].set_title('Power = 2')

axs[1,0].plot(x_data, x_data**3, label='Power = 3', color='brown')
axs[1,0].set_title('Power = 3')

axs[1,1].plot(x_data, x_data**4, label='Power = 4', color='orange')
axs[1,1].set_title('Power = 4')

fig.tight_layout()
```
